```python
# visualisasi grafik
import numpy as np
import matplotlib.pyplot as plt
ys = 200 + np.random.randn(100)
x = [i for i in range(len(ys))]
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.show()

plt.title("Sample Visualization")
plt.show()
```
