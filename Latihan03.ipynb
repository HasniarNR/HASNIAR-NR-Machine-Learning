```python
#contoh if dua kondisi
angka = int(input('Masukkan angka: '))
if (angka % 2 == 0):
   print('Angka genap terdeteksi')
   print('%d adalah angka genap' % (angka))
else:
   print('Angka ganjil terdeteksi')
   print('%d adalah angka ganjil' % (angka))
```

```python
#contoh penggunaan while dan break
i = 0
while (i <= 6):
   print(i)
   i += 1
   break #break setelah proses increment

print('Pesan ini di luar perulangan')
```
