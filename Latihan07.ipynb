```python
# prompt: buatkan scrip untun menghitung persegi panjang

def hitung_persegi_panjang(panjang, lebar):
  """
  Hitung luas persegi panjang.

  Args:
    panjang: Panjang persegi panjang dalam satuan cm.
    lebar: Lebar persegi panjang dalam satuan cm.

  Returns:
    Luas persegi panjang dalam satuan cm^2.
  """
  luas = panjang * lebar
  return luas

# Example usage
panjang = 10
lebar = 5

luas = hitung_persegi_panjang(panjang, lebar)

print("Luas persegi panjang:", luas, "cm^2")
```
